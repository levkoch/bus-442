```python
# Imports
import os
import pandas as pd
import numpy as np
from janitor import clean_names
import calendar
import datetime
```

```python
overall_df = pd.DataFrame()
```

```python
for idx in range(1, 13):
  days = calendar.monthrange(2023, idx)[1]
  for i in range(1, days + 1):
   name = calendar.month_name[idx]
   file_path = "./data/" + name + "/results_for_2023-" + "{:02n}".format(idx) + "-" + "{:02n}".format(i) +".csv"
   print(file_path)
   df = clean_names(pd.read_csv(file_path))
   df = df[['operation_date','service_rte_num','patt_direction','stop_id',
    'first_act_dep_time', 'first_act_arr_time', 'last_act_arr_time', 'last_act_dep_time', 'sched_arr_time', 'sched_dep_time', 'psngr_load']]
   df = df[df['sched_arr_time'] > 0]
   overall_df = pd.concat([overall_df, df])
```

```python
# Calulcate lateness for all columns
overall_df['lateness'] = overall_df['first_act_arr_time'] - overall_df['sched_arr_time']
overall_df = overall_df[overall_df['lateness'] > -3600]
overall_df = overall_df[overall_df['lateness'] < 5400]
```

```python
# Add days of week column
day_names = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
overall_df['weekday'] = [datetime.datetime.strptime(overall_df['operation_date'].iloc[i], '%Y-%m-%d').strftime('%A') for i in range(len(overall_df))]
```
