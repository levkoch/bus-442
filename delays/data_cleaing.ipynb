```python
# Imports
import os
import pandas as pd
import numpy as np
from janitor import clean_names
import calendar
import datetime
```

```python
overall_df = pd.DataFrame()
```

```python
for idx in range(1, 13):
  days = calendar.monthrange(2023, idx)[1]
  for i in range(1, days + 1):
   name = calendar.month_name[idx]
   file_path = "./data/" + name + "/results_for_2023-" + "{:02n}".format(idx) + "-" + "{:02n}".format(i) +".csv"
   print(file_path)
   df = clean_names(pd.read_csv(file_path))
   df = df[['operation_date','service_rte_num','patt_direction','stop_id',
    'first_act_dep_time', 'first_act_arr_time', 'last_act_arr_time', 'last_act_dep_time', 'sched_arr_time', 'sched_dep_time', 'psngr_load']]
   df = df[df['sched_arr_time'] > 0]
   overall_df = pd.concat([overall_df, df])
```

```python
# Renaming:
df = pd.read_csv("AverageDaysFormatted.csv")
df = df.drop('Unnamed: 0', axis=1)
```

```python
df[df['lateness'] < 0].head()
```

```python
df['lateness minutes'] = pd.to_datetime(abs(df['lateness']), unit='s').dt.strftime('%M:%S')
```

```python
# Convert weekday to indicies
DAY_COLS = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day_arr'] = [DAY_COLS.index(i) for i in df['weekday']]
df = df.drop('weekday', axis=1)
df = df.drop('psngr_load', axis=1)
```

```python
df.head()
```

```python
df_lateness = df.groupby(['servi_numce_rte', 'patt_direction', 'stop_id', 'sched_arr_time', 'sched_arr_time_formatted'])['lateness'].apply(list).reset_index(name='lateness')
df_latemins = df.groupby(['servi_numce_rte', 'patt_direction', 'stop_id', 'sched_arr_time', 'sched_arr_time_formatted'])['lateness minutes'].apply(list).reset_index(name='lateness_minutes')
df_dayidx = df.groupby(['servi_numce_rte', 'patt_direction', 'stop_id', 'sched_arr_time', 'sched_arr_time_formatted'])['day_arr'].apply(list).reset_index(name='day_arr')
df = df.drop('lateness', axis=1)
df = df.drop('lateness minutes', axis=1)
df = df.drop('day_arr', axis=1)
df = df.merge(
    df_lateness[['servi_numce_rte', 'patt_direction', 'stop_id', 'sched_arr_time', 'sched_arr_time_formatted', 'lateness']],
    on=['servi_numce_rte', 'patt_direction', 'stop_id', 'sched_arr_time', 'sched_arr_time_formatted'], how='left'
).merge(
    df_latemins[['servi_numce_rte', 'patt_direction', 'stop_id', 'sched_arr_time', 'sched_arr_time_formatted', 'lateness_minutes']],
    on=['servi_numce_rte', 'patt_direction', 'stop_id', 'sched_arr_time', 'sched_arr_time_formatted'], how='left'
).merge(
    df_dayidx[['servi_numce_rte', 'patt_direction', 'stop_id', 'sched_arr_time', 'sched_arr_time_formatted', 'day_arr']],
    on=['servi_numce_rte', 'patt_direction', 'stop_id', 'sched_arr_time', 'sched_arr_time_formatted'], how='left'
)
df.head()
```

```python
# Drop duplicates
df = df.drop_duplicates(subset=["servi_numce_rte", "patt_direction", 'stop_id', 'sched_arr_time', "sched_arr_time_formatted"])
```
